# The Simple OLS Regression Model

The simple regression model can be used to study the relationship between two variables. In this notebook, we will examine the relationship between education and wage.

We will import the necessary libraries and load the wage1 dataset from the wooldridge package. This dataset contains information on wages, education, and other characteristics of workers.

```python
import numpy as np
import statsmodels.formula.api as smf
import wooldridge as woo
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid', palette='muted')
```

```python
wage = woo.dataWoo('wage1')
wage.head()
```

```python
results = smf.ols('np.log(wage) ~ educ', data=wage).fit()

b = results.params

print(f"wage_hat = {b.iloc[0]:.2f} + {b.iloc[1]:.2f} educ")
```

```python
sns.regplot(x='educ', y='wage', data=wage, scatter_kws={'alpha':0.5}, line_kws={'color':sns.color_palette()[1]})

plt.title('Wage vs. Education')
plt.xlabel('Years of Education')
plt.ylabel('log(Wage)')

plt.grid(alpha=0.2)

sns.despine()
```

Based on this sample data, we can see that there is a positive relationship between education and wage. Each additional year of education is associated with an 8% increase in wage, on average.

Importantly, we have not proven a causal relationship between education and wage. There could be other factors that influence both education and wage, such as ability, family background, or access to resources. We would need more advanced econometric techniques to establish causality—which we will cover later.
